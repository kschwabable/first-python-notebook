Now we are going to practice cleaning up a sloppy dataset, where Sean Parker's name appears twice. 

```python
def combine_names(row):
    if row.contributor_fullname.startswith('SEAN PARKER'):
        return 'SEAN PARKER'
    return row.contributor_fullname
```

```python
top_supporters.apply(combine_names, axis=1)
```

```python
top_supporters['contributor_cleanname'] = top_supporters.apply(combine_names, axis=1)
```

```python
top_supporters.groupby("contributor_cleanname").amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
clean_supporters = top_supporters.groupby("contributor_cleanname").amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
chart = clean_supporters.head(5).amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_cleanname)
```
